```python
import re
import nltk
import gensim.downloader as api
wv = api.load('word2vec-google-news-300')
```

```python
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
lem = WordNetLemmatizer()
```

```python
df['REVIEW']=None
for i in range(0,len(df)):
    review = re.sub('[^a-zA-Z0-9]',' ',df['CONTENT'][i])
    review = review.lower()
    review = review.split()
    
    review = [lem.lemmatize(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    df['REVIEW'][i] = review
```

```python
corpus=df['REVIEW']
corpus
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
```

```python
nltk.download('punkt')
```

### Making our own dictionary of words

```python
words=[]
for sent in corpus:
    sent_token = sent_tokenize(sent)
    for sent in sent_token:
        words.append(simple_preprocess(sent))
```

```python
df[df['REVIEW']==""]
```

```python
len(words)
```

```python
import gensim
model = gensim.models.Word2Vec(words,window=7,min_count=0)
```

```python
def avgword2vec(doc):
    return np.mean([model.wv[word] for word in doc if word in model.wv.index_to_key], axis=0)
```

```python
from tqdm import tqdm
X=[]
for i in tqdm(range(0,len(words))):
    X.append(avgword2vec(words[i]))
```
